```python
# Define variables
x, y, z = sp.symbols('x y z')

# Question 1: Find the equation of the plane passing through points A(1, 2, 3), B(3, 4, 5), and C(2, 1, 4)
A = sp.Matrix([1, 2, 3])
B = sp.Matrix([3, 4, 5])
C = sp.Matrix([2, 1, 4])

# Create vectors
AB = B - A
AC = C - A

# Compute the normal vector using cross product
normal_vector = AB.cross(AC)

# Equation of the plane
plane_eq1 = normal_vector[0] * (x - A[0]) + normal_vector[1] * (y - A[1]) + normal_vector[2] * (z - A[2])

print("Question 1 - Equation of the plane:")
print(sp.simplify(plane_eq1), "= 0")

# Question 2: Find the equation of a plane passing through A(1, 2, 3) and parallel to 2x + 3y + 4z = 5
point = sp.Matrix([1, 2, 3])
normal_parallel = sp.Matrix([2, 3, 4])  # Normal vector of the parallel plane
plane_eq2 = normal_parallel[0] * (x - point[0]) + normal_parallel[1] * (y - point[1]) + normal_parallel[2] * (z - point[2])

print("\nQuestion 2 - Equation of the parallel plane:")
print(sp.simplify(plane_eq2), "= 0")

# Question 3: Find the line of intersection between 2x + 3y + 4z = 5 and 3x + y + 2z = 6
plane1 = 2 * x + 3 * y + 4 * z - 5
plane2 = 3 * x + y + 2 * z - 6

# Compute the direction vector of the intersection line
normal1 = sp.Matrix([2, 3, 4])
normal2 = sp.Matrix([3, 1, 2])
direction_vector = normal1.cross(normal2)

# Solve for a point on the intersection line
solution_point = sp.linsolve([plane1, plane2], (x, y, z))

print("\nQuestion 3 - Direction vector of the intersection line:")
print(direction_vector)

print("A point on the intersection line:")
print(solution_point)

# Question 4: Find the equation of a plane passing through A(1, 2, 3) and perpendicular to 2x + 3y + 4z = 5
plane_eq4 = normal_parallel[0] * (x - A[0]) + normal_parallel[1] * (y - A[1]) + normal_parallel[2] * (z - A[2])

print("\nQuestion 4 - Equation of the perpendicular plane:")
print(sp.simplify(plane_eq4), "= 0")

# Question 5: Find the plane passing through points A(2, 0, 0), B(0, 3, 0), C(0, 0, 4), and calculate the angle with x+y+z=1
A5 = sp.Matrix([2, 0, 0])
B5 = sp.Matrix([0, 3, 0])
C5 = sp.Matrix([0, 0, 4])

AB5 = B5 - A5
AC5 = C5 - A5

# Compute the normal vector for the plane
normal_vector5 = AB5.cross(AC5)
plane_eq5 = normal_vector5[0] * (x - A5[0]) + normal_vector5[1] * (y - A5[1]) + normal_vector5[2] * (z - A5[2])

# Calculate the angle between the planes
normal5 = normal_vector5 / sp.sqrt(normal_vector5.dot(normal_vector5))
normal_xy = sp.Matrix([1, 1, 1]) / sp.sqrt(3)
angle = sp.acos(normal5.dot(normal_xy)) * 180 / sp.pi

print("\nQuestion 5 - Equation of the plane:")
print(sp.simplify(plane_eq5), "= 0")

print("Angle between the planes:")
print(angle.evalf(), "degrees")
```
