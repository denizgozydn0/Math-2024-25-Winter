```python
# Sembol tanımlamaları
x, y = sp.symbols('x y')

# 1. A(1, 2) ve B(3, 4) noktalarından geçen doğru denklemi
slope_AB = (4 - 2) / (3 - 1)  # Eğimi hesapla
line_AB = sp.Eq(y, slope_AB * (x - 1) + 2)

# 2. A(1, 2) noktasından geçen ve y = 2x + 3'e paralel doğru denklemi
slope_parallel = 2
line_parallel = sp.Eq(y, slope_parallel * (x - 1) + 2)

# 3. A(1, 2) noktasından geçen ve y = 2x + 3'e dik doğru denklemi
slope_perpendicular = -1 / slope_parallel  # Dik doğru eğimi
line_perpendicular = sp.Eq(y, slope_perpendicular * (x - 1) + 2)

# 4. y = 2x + 3 ve y = 3x + 2 doğrularının kesişim noktası
line1 = sp.Eq(y, 2 * x + 3)
line2 = sp.Eq(y, 3 * x + 2)
intersection = sp.solve([line1, line2], (x, y))

# 4. Doğrular arasındaki açı
angle_between_lines = math.degrees(math.atan(abs((3 - 2) / (1 + 3 * 2))))

# 5. A(1, 2) noktasından geçen ve [2, 3] vektörüne paralel doğru denklemi
vector_parallel = [2, 3]
slope_vector = vector_parallel[1] / vector_parallel[0]  # Vektör eğimi
line_vector = sp.Eq(y, slope_vector * (x - 1) + 2)

# 6. A(1, 2) noktasının y = 2x + 3 doğrusuna olan uzaklığı
distance = abs(2 * 1 - 2 + 3) / math.sqrt(2**2 + 1)

# 7. y = 2x + 3 doğrusunun eksenlerle kesişim noktaları
intersection_x = sp.solve(line1.subs(y, 0), x)  # x-eksenini kestiği nokta
intersection_y = sp.solve(line1.subs(x, 0), y)  # y-eksenini kestiği nokta

# 8. y = x + 3 doğrusu ile x-ekseni arasındaki açı
angle_with_x_axis = math.degrees(math.atan(1))  # Eğimi 1 olan doğrunun açısı

# 9. x + y + 1 = 0 doğrusuna dik vektör
vector_perpendicular = [1, 1]  # Doğru x + y + 1 = 0'a dik olan vektör

# Sonuçları yazdır
print("1. Line through A and B:", line_AB)
print("2. Line parallel to y=2x+3 through A:", line_parallel)
print("3. Line perpendicular to y=2x+3 through A:", line_perpendicular)
print("4. Intersection of y=2x+3 and y=3x+2:", intersection)
print("5. Angle between y=2x+3 and y=3x+2:", angle_between_lines, "degrees")
print("6. Line through A parallel to vector [2,3]:", line_vector)
print("7. Distance from A to y=2x+3:", distance)
print("8. Intersection with axes (x-axis, y-axis):", (intersection_x, intersection_y))
print("9. Angle with x-axis:", angle_with_x_axis, "degrees")
print("10. Perpendicular vector to x+y+1=0:", vector_perpendicular)
```
